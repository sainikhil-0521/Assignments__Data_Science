Question-1

Matplotlib is a powerful Python library used for creating static, animated, and interactive visualizations. It provides a wide range of functions and classes for generating various types of plots and charts.

Matplotlib is used for data visualization, which is the process of representing data in a graphical or visual format. Visualization helps in understanding patterns, trends, and relationships within the data, making it easier to draw insights and communicate findings effectively.

1 Line Plot:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 15, 11]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

2 Bar Plot:

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 35]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

3. Scatter Plot:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 15, 11]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

4 Histogram:

```python
import matplotlib.pyplot as plt

data = [1, 3, 2, 2, 4, 3, 5, 3, 2, 1]

plt.hist(data, bins=5)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

5 Pie Chart:

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [30, 25, 20, 25]

plt.pie(sizes, labels=labels)
plt.title('Pie Chart')
plt.show()
```

Question-2

A scatter plot is a type of plot that displays individual data points as markers on a 2D plane. It is used to visualize the relationship between two variables. Each point on the scatter plot represents the values of the two variables for a specific data point.



```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Question-3

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to display multiple plots or charts in a single figure window, making it easier to compare and analyze different datasets or variables.

The subplot() function takes three arguments: subplot(nrows, ncols, index). Here, nrows and ncols specify the number of rows and columns in the subplot grid, and index indicates the position of the current subplot within the grid.

```python
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line Plot 1')


y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line Plot 2')

y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line Plot 3')


y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line Plot 4')

plt.tight_layout() 
plt.show()
```

Question-4

A bar plot, also known as a bar chart, is a type of plot that uses rectangular bars to represent data. It is commonly used to compare and display categorical data, where the length or height of each bar represents the value or frequency of a category.

Bar plots are used to visualize and analyze categorical variables and their relationships. They allow for easy comparison between different categories, identification of patterns or trends, and highlighting differences or similarities in the data.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.figure(figsize=(8, 6))
plt.bar(company, profit)
plt.xlabel('company')
plt.ylabel('profit')
plt.title('Bar Plot')
plt.show()

plt.figure(figsize=(8, 6))
plt.barh(company, profit)
plt.xlabel('profit')
plt.ylabel('company')
plt.title('Barh Plot')
plt.show()
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the summary statistics, such as the minimum, first quartile (25th percentile), median (second quartile), third quartile (75th percentile), and maximum values, as well as any potential outliers.

Box plots are used to visualize the distribution and variability of numerical data. They provide insights into the range, skewness, symmetry, and potential outliers in the dataset. Box plots are especially useful for comparing multiple groups or categories side by side.



```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.figure(figsize=(8, 6))
plt.boxplot(box1)
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Box Plot 1')
plt.show()

plt.boxplot(box2)
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Box Plot 2')
plt.show()
```
